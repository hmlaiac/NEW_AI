# Part A. Image Analysis with Human Eye

```python
import cv2 
import numpy as np  

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("img\\go_board.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

## A1. Convert the image to Gray Scale

Gray scale is a color with light intensity from 0 to 255. Computer can identify the strength of pixels easily

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap="gray")
```

## A2. Data Observation

```python
print("Highest intensity of the go board is: %d"%(gray_img.max()))
print("Lowest intensity of the go board is: %d"%(gray_img.min()))
```

```python
dst = cv2.cornerHarris(src=gray_img,blockSize=2,ksize=3,k=0.04)
dst = cv2.dilate(dst,None)
```

```python
img_copy = img.copy()
img_copy[dst>0.3*dst.max()]=[255,0,0]

plt.imshow(img_copy)
```

# Part B. All in one code driven by widegets

```python
import ipywidgets as widgets
from ipywidgets import HBox, VBox
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import cv2 
import numpy as np  
%matplotlib inline
```

```python
@widgets.interact(x=(0, 1.00000), blockSize=(1,10), ksize=(3,30))
def f(x=0, blockSize=1, ksize=3):
    if ksize%2==0:
        ksize+=1
    dst = cv2.cornerHarris(src=gray_img,blockSize=blockSize,ksize=ksize,k=0.04)
    dst = cv2.dilate(dst,None)
    img_copy = img.copy()
    img_copy[dst>x*dst.max()]=[255,0,0]

    plt.imshow(img_copy)
```
