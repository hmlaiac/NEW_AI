## Part B1. Encode type with LabelEncoder

```python
type_val = df['type'].unique()
le_type = preprocessing.LabelEncoder()
le_type.fit(type_val)
df['type'] = le_type.transform(df['type'])
```

## Part B2. T-test for data

In mathematics, you can use statistical methods to analyze sales data and determine if a sale period existed. Here are some steps you can take to prove the existence of a sale period using mathematical methods:
1. f the difference between the means is statistically significant (usually with a p-value less than 0.05), then you can reject the null hypothesis and conclude that a sale period existed.
2. The null hypothesis is that there was no sale period, and any observed increase in sales was due to chance. The alternative hypothesis is that there was a sale period, and the increase in sales was due to the sale.

```python
import pandas as pd
from scipy.stats import ttest_ind

# Create two sample datasets with different shapes
sample1 = selected_data1["Total Sale"]
sample2 = selected_data2["Total Sale"]
sample3 = selected_data3["Total Sale"]

# Conduct a t-test on the two samples with unequal variances
t_statistic, p_value = ttest_ind(sample1, sample2 ,equal_var=False)
```
