# Different Signal Graphs

## 1. Convolutions

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 20)
y = np.sin(x)

plt.stem(y)
plt.plot(y)
plt.show()
```

```python
y1 = np.array([3])
y2 = np.convolve(y, y1)
plt.stem(y2)
plt.plot(y2)
plt.show()
```

# Part A. Project Set Up
- The sampling rate is 50 Hz
- Most of time should only show 1 Hz signal to easy to identify the result of convolution

```python
# Generate data
t = np.linspace(0, 1, 100)
f = 1 # 5Hz = 0.2s
y = np.abs(2*f*t - 2*np.floor(f*t + 0.5))

# Plot the data
plt.plot(t, y)

# Customize the plot
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Triangle Wave')
plt.grid(True)

# Display the plot
plt.show()
```

## A1. Sine wave

```python
# Define the frequency and time range
f = 1 # frequency in Hz

t = np.linspace(0, 1, 50) # time range in seconds

# Generate the sine wave
y = np.sin(2*np.pi*f*t)

# Plot the data
plt.plot(y)
plt.stem(y)
# Customize the plot
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave with f={} Hz'.format(f))
plt.grid(True)

# Display the plot
plt.show()
```

```python
y1 = np.array([0,1,1,1,0,1,1,1,0])

# Plot the data
plt.stem(y1)
# Customize the plot
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine Wave with f={} Hz'.format(f))
plt.grid(True)

# Display the plot
plt.show()
```

```python
y2 = np.convolve(y, y1)
plt.stem(y2)
plt.plot(y2)
plt.show()
```
