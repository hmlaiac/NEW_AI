```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("img\\steve.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
plt.imshow(edges, cmap='gray_r')
```

## Part B. Chossing Threshold

```python
median_val = np.median(img)
```

```python
print("The maximum intensity of the image %d"%img.max())
print("The minimum intensity of the image %d"%img.min())
```

```python
const1, const2 = 0.7, 1.3
lower = int(max(0, const1* median_val))
upper = int(min(255,const2 * median_val))
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper)
plt.imshow(edges, cmap='gray_r')
```

## Part C. Data Analysis with blurred image

```python
blurred_img = cv2.blur(img,ksize=(5,5))

const1, const2 = 0.1, 0.5
lower = int(max(0, const1* median_val))
upper = int(min(255,const2 * median_val))

edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper)
plt.imshow(edges, cmap='gray_r')
```
