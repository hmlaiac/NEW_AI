## A. Image Processing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### A1. Original Image

```python
im = cv2.imread("data\\png_ball1.png")
im_original = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.title("Original Image")
plt.imshow(im_original)
plt.show()
```

### A2. Color Detection

```python
im_hsv = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
# take only the orange, highly saturated, and bright parts
# im_hsv = cv2.inRange(im_hsv, (7, 180, 180), (11, 255, 255))
im_hsv = cv2.inRange(im_hsv, (0, 150, 100), (20, 255, 200))

# To show the detected orange parts:
im_orange = im.copy()
im_orange[im_hsv == 0] = 0
plt.imshow(cv2.cvtColor(im_orange, cv2.COLOR_BGR2RGB))
plt.plot()
```

The main object is detected but we need to locate it accurately, so we apply gray and blur image to remove noise surrounding object and try to use erode and diate method to make the main object more clear

```python
im_gray = cv2.cvtColor(im_orange, cv2.COLOR_BGR2GRAY)
ksize = (10, 10)
im_gray_blur = cv2.blur(im_gray, ksize)
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(im_gray, cmap="gray")
plt.subplot(122)
plt.imshow(im_gray_blur, cmap="gray")
plt.show()
```

```python
kernel = np.ones((21,21), np.uint8)
erosion = cv2.erode(im_gray_blur, kernel, iterations = 1)
plt.imshow(im_gray_blur, cmap="gray")
plt.show()
plt.imshow(erosion, cmap="gray")
plt.show()
```

```python
kernel = np.ones((11,11), np.uint8)
opening = cv2.morphologyEx(erosion, cv2.MORPH_OPEN, kernel)
plt.imshow(opening, cmap="gray")
plt.show()
```

```python
kernel = np.ones((21,21), np.uint8)
dilate = cv2.dilate(erosion, kernel, iterations = 1)
plt.imshow(erosion, cmap="gray")
plt.show()
plt.imshow(dilate, cmap="gray")
plt.show()
```

### A3. Shape Detection

```python
im_copy = im.copy()
circles = cv2.HoughCircles(dilate, cv2.HOUGH_GRADIENT, 1, 50,
                                   param1=50, param2=30, minRadius=10, maxRadius=300)

if circles is not None:
    circles = np.round(circles[0, :]).astype("int")
    for (x, y, r) in circles:

        cv2.circle(im_copy, (x, y), r, (0, 255, 0), 2)

# Display the processed frame
# cv2.imshow('Ball Detection', img)
plt.imshow(cv2.cvtColor(im_copy, cv2.COLOR_BGR2RGB))
```

## B. Testing

## Build a function to detect image

```python
def basket_ball_detect(im):
    # Color Filtering
    im_hsv = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
    im_hsv = cv2.inRange(im_hsv, (0, 150, 100), (20, 255, 200))
    im_orange = im.copy()
    im_orange[im_hsv == 0] = 0

    # Convert to gray image
    im_gray = cv2.cvtColor(im_orange, cv2.COLOR_BGR2GRAY)
    
    # Blur Image
    ksize = (10, 10)
    im_gray_blur = cv2.blur(im_gray, ksize)
        
    # Erosion
    kernel = np.ones((11,11), np.uint8)
    erosion = cv2.erode(im_gray_blur, kernel, iterations = 1)
    
    #Dialate
    kernel = np.ones((21,21), np.uint8)
    dilate = cv2.dilate(erosion, kernel, iterations = 1)
    
    # Closing
    kernel = np.ones((21,21), np.uint8)
    closing = cv2.morphologyEx(dilate, cv2.MORPH_CLOSE, kernel)
    
    plt.imshow(erosion, cmap="gray")
    plt.show()
    plt.imshow(dilate, cmap="gray")
    plt.show()
    plt.imshow(closing, cmap="gray")
    plt.show()
    
    # Circle detect
    im_copy = im.copy()
    circles = cv2.HoughCircles(closing, cv2.HOUGH_GRADIENT, 1, 200,
                                       param1=50, param2=30, minRadius=50, maxRadius=300)

    if circles is not None:
        circles = np.round(circles[0, :]).astype("int")
        for (x, y, r) in circles:

            cv2.circle(im_copy, (x, y), r, (0, 255, 0), 2)
    return im_copy
```

```python
# Read image
img = cv2.imread("data\\ball_3.jpg")
out = basket_ball_detect(img)
plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB))
```

```python
# Read image
img = cv2.imread("data\\png_ball1.png")
out = basket_ball_detect(img)
plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB))
```
